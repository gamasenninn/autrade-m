```python

import yfinance as yf
import matplotlib.pyplot as plt
import datetime
import pandas as pd

positions =[]
positions_log =[]
commits = []

#df_positions =[]
#df_positions_log =[]
df_commits = pd.DataFrame()

stocker = "5020.T"
qty = 200
#deal_max = -200
deal_start = "2022-01-01"
deal_end = "2022-12-31"
eval_rate = 3.0
commit_rate = 6.0
max_position = 5
loss_cut_rate = -100.0
max_duration = 300

is_min_buy = True
is_min_sell = True
is_max_buy = True
is_max_sell = True

weight ={
  'min_buy' : 20,
  'min_sell' : 10,
  'max_buy' : 5,
  'max_sell' : 30,    
}

is_buy = True
is_sell = True
#-------- ポジションの判定 -----
def eval_each_position(row,deal_type):
  if len(positions) < max_position:
    min_price = 999999999999
    max_price = 0
    cur_price = float(row['Close'])
    #eval_price = cur_price * qty
    for pos in positions:
      p_price = pos['unit_price']
      if pos['type'] == deal_type:
        if p_price <= min_price:
            min_price = p_price 
        if p_price >= max_price:
            max_price = p_price 
    print (f"min-max({deal_type}): min={min_price} , max={max_price} ,current={cur_price}")
    if deal_type == "buy":
      if is_max_buy:
        if (max_price+max_price*eval_rate/100) < cur_price:
          print (f"hit max({deal_type}): max={max_price} ,current={cur_price}")
          return True
      if is_min_buy:
        if (min_price-min_price*eval_rate/100) > cur_price:
          print (f"hit min({deal_type}): max={min_price} ,current={cur_price}")
          return True      
    if deal_type == "sell":
      if is_max_sell:
        if (max_price+max_price*eval_rate/100) < cur_price:
          print (f"hit max({deal_type}): max={max_price} ,current={cur_price}")
          return True
      if is_min_sell:
        if (min_price-min_price*eval_rate/100) > cur_price:
          print (f"hit min({deal_type}): max={min_price} ,current={cur_price}")
          return True      
  return False

def check_available_position(deal_type):
  count = 0
  for pos in positions:
    if pos['type'] == deal_type:
      count += 1
  if count < max_position-1:
    return True
  return False

def eval_position(row):
  #初期状態
  if  len(positions) == 0:
    if is_buy: 
      return "buy"
    else:
      return "sell"

  deal_types = ["sell","buy"]

  if eval_each_position(row,"sell") \
    and is_sell \
    and check_available_position("sell"):
    return "sell"


  if eval_each_position(row,"buy") \
    and is_buy \
    and check_available_position("buy"):
    return "buy"


  return ""
#-------- ポジションを持つ -----
def entry_position(idx,row,type):
  price = int(float(row['Close'])*qty)
  pl = {
      'deal_date': idx.strftime('%Y-%m-%d'),
      'type': deal_type,
      'unit_price': float(row['Close']),
      'qty': qty,
      'price': price
  }
  print("entry:",pl)
  positions.append(pl)
  positions_log.append(pl)

#-------- ポジションの再計算　--------
def calc_position():
  for pos in positions:
    current_price = float(row['Close'])
    eval_price = int(current_price * qty)
    price = pos['price']
    profit = int(eval_price - price)
    reverse = -1 if pos['type'] == "sell" else 1
    pos['current_price'] = current_price
    pos['eval_price'] = eval_price
    pos['profit'] = profit * reverse
    pos['p_rate'] = (1-price/eval_price)*100 *reverse
    pos['duration'] = pos['duration'] +1 if 'duration' in pos else 0
    if 'max_down_rate' in pos:
      if pos['max_down_rate'] > pos['p_rate']:
         pos['max_down_rate'] = pos['p_rate']
    else:
        pos['max_down_rate'] = pos['p_rate']
    pos['commit'] = False
 
    #print("再計算:",pos)

#-------- ポジションの決済　--------
def check_available_commit(deal_type):
  return True
  #--- なんかうまくいかない
  deal_count = 0
  for pos in positions:
    if pos['type'] == deal_type:
      deal_count += 1

  if len(positions)>0 and deal_count==1:
    return False
  return True

def commit(pos):
  positions.remove(pos)
  commits.append(pos)
  global df_commits 
  df_commits = df_commits.append(pos,ignore_index=True)

def commit_by_batch():

  pos_l = [pos for pos in positions if pos['commit']]

  for pos in pos_l:
    if pos['commit']:
      if pos['profit'] < 0:
        print("-相殺実行:",pos)
        #positions.remove(pos)
        commit(pos)
      else:
        print("+相殺実行",pos) 
        #positions.remove(pos)
        commit(pos)
      #commits.append(pos)

def setoff_commit():
  plus_count = 0
  plus_price = 0
  is_setoff = False
  for pos in positions:
    if pos['profit'] > 0:
      plus_count += 1
      plus_price += pos['profit'] 
  if plus_count >= 2:
    print("相殺決済可能:",plus_price)
    for idx,pos in enumerate(positions):
      if pos['profit']<0 and abs(pos['profit']) < plus_price:
        #print("-相殺実行:")
        pos['commit'] = True
        #commit_by_index(idx)
        is_setoff = True
        break
    if is_setoff:
      for idx,pos in enumerate(positions):
        if pos['profit'] > 0:
          #print("＋相殺実行:")
          pos['commit'] = True
      [print(pos) for pos  in positions]
      commit_by_batch()
      return True

  return False



def commit_by_index(idx):
  pos = positions[idx]
  print("\ncommit:",pos)
  commit(pos)
  #commits.append(pos)
  #positions.pop(idx)
def commit_by_index_list(idx_l):
  for idx in idx_l:
    pos = positions[idx]
    print("\ncommit:",pos)
    commit(pos)
    #commits.append(pos)
  #positions.pop(idx_l)

def commit_position():
  print(len(positions),end=",")
  if setoff_commit():
    pass
    #return True
  #--- 通常決済 ----
  for i, pos in enumerate(positions):
    if pos['p_rate'] > commit_rate:
      if check_available_commit(pos['type']):
        commit_by_index(i)
        break

  #--- 強制ロスカット ----
  for i, pos in enumerate(positions):
    if pos['p_rate'] < loss_cut_rate:
      if check_available_commit(pos['type']):
        commit_by_index(i)
        break

  #--- 時限ロスカット ----
  for i, pos in enumerate(positions):
    if pos['duration'] > max_duration:
      if check_available_commit(pos['type']):
        commit_by_index(i)
        break

  return True

#-------- 取引シミュレーション -----
#ticker_info = yf.Ticker("5020.T")
ticker_info = yf.Ticker(stocker)
hist = ticker_info.history(start=deal_start,end=deal_end)
plt.plot(hist['Close'])
#plt.plot(df_commits['profit'].cumsum())

for idx,row in hist.iterrows():
  calc_position()
  deal_type = eval_position(row)
  if deal_type:
      entry_position(idx,row,deal_type)
  calc_position()
  commit_position()

print("\ncommits:")
#----- 最終計
commit_total = 0
df_commits['cumsum'] = df_commits['profit'].cumsum()
pd.set_option('display.expand_frame_repr', False)
print(df_commits)
for commit in commits:
  commit_total += commit['profit']
  #print(commit)
print("\ntotal:",commit_total, len(commits))

position_total = 0
for position in positions:
  print("psition:",position)
  position_total += position['profit']
print("\npsition total:",position_total, len(positions))

print("\nfinal:", commit_total + position_total)

#df_commits.fillna(0)
#plt.plot(df_commits['profit'].cumsum())

```

相殺ロジックで降順でcheckするように。そうすることで大きいポジションから相殺するようにできる。  
しかし結論からいうと、あまり変化はない。最大期待ポジションは解消されにくい。
また、ドローダウンをcheckできるように、ポジションサマリーテーブルも付与する。
ポジションバランスにしたがって、売買パラメータを動的に決定するロジック付与する。


```python

import yfinance as yf
import matplotlib.pyplot as plt
import datetime
import pandas as pd

positions =[]

df_commits = pd.DataFrame(columns=['date','deal_date','type','current_price','price','unit_price','qty','eval_price','profit'])
df_positions_log =pd.DataFrame()
df_positions_sum_log =pd.DataFrame()

stocker = "5020.T" 
#stocker = "5108.T" 
qty = 200
#deal_max = -200
deal_start = "2011-01-01"
deal_end = "2011-12-31"
eval_rate = 3.0
commit_rate = 6.0
max_position = 5
loss_cut_rate = -100.0
max_duration = 300

is_min_buy = True
is_min_sell = True
is_max_buy = True
is_max_sell = True

weight ={
  'min_buy' : 20,
  'min_sell' : 10,
  'max_buy' : 5,
  'max_sell' : 30,    
}

is_buy = True
is_sell = True

_ROW = []

#---- 実験パラメータ ------
sell_danger_line = -5000000000
buy_danger_line = -50000000000
#-------- ポジションの判定 -----
def eval_each_position(row,deal_type):
  if len(positions) < max_position:
    min_price = 999999999999
    max_price = 0
    cur_price = float(row['Close'])
    #eval_price = cur_price * qty
    for pos in positions:
      p_price = pos['unit_price']
      if pos['type'] == deal_type:
        if p_price <= min_price:
            min_price = p_price 
        if p_price >= max_price:
            max_price = p_price 
    print (f"min-max({deal_type}): min={min_price} , max={max_price} ,current={cur_price}")
    if deal_type == "buy":
      if is_max_buy:
        if (max_price+max_price*eval_rate/100) < cur_price:
          print (f"hit max({deal_type}): max={max_price} ,current={cur_price}")
          return True
      if is_min_buy:
        if (min_price-min_price*eval_rate/100) > cur_price:
          print (f"hit min({deal_type}): max={min_price} ,current={cur_price}")
          return True      
    if deal_type == "sell":
      if is_max_sell:
        if (max_price+max_price*eval_rate/100) < cur_price:
          print (f"hit max({deal_type}): max={max_price} ,current={cur_price}")
          return True
      if is_min_sell:
        if (min_price-min_price*eval_rate/100) > cur_price:
          print (f"hit min({deal_type}): max={min_price} ,current={cur_price}")
          return True      
  return False

def check_available_position(deal_type):
  count = 0
  for pos in positions:
    if pos['type'] == deal_type:
      count += 1
  if count < max_position-1:
    return True
  return False

def eval_even_position():

  buy_count = 0
  sell_count = 0
  for pos in positions:
    if pos['type'] == 'buy':
      buy_count += 1
    else:
      sell_count += 1

  if buy_count+sell_count >= max_position:
    return ""

  if buy_count > sell_count:
    return "sell"
  elif sell_count > buy_count:
    return "buy"
  else:
    return ""

def eval_position(row):
  #初期状態
  if  len(positions) == 0:
    if is_buy: 
      return "buy"
    else:
      return "sell"

  deal_types = ["sell","buy"]

  eval_pos = eval_even_position() # even position
  if eval_pos:
    return eval_pos

  if eval_each_position(row,"sell") \
    and is_sell \
    and check_available_position("sell"):
    return "sell"


  if eval_each_position(row,"buy") \
    and is_buy \
    and check_available_position("buy"):
    return "buy"


  return ""
#-------- ポジションを持つ -----
def entry_position(idx,row,type):
  price = int(float(row['Close'])*qty)
  pl = {
      'deal_date': idx.strftime('%Y-%m-%d'),
      'type': deal_type,
      'unit_price': float(row['Close']),
      'qty': qty,
      'price': price
  }
  print("entry:",pl)
  positions.append(pl)
  #positions_log.append(pl)
  global df_positions_log 
  df_positions_log = df_positions_log.append(pl,ignore_index=True)

#-------- ポジションの再計算　--------
def calc_position(idx,row):
  for pos in positions:
    current_price = float(row['Close'])
    eval_price = int(current_price * qty)
    price = pos['price']
    profit = int(eval_price - price)
    reverse = -1 if pos['type'] == "sell" else 1
    pos['current_price'] = current_price
    pos['eval_price'] = eval_price
    pos['profit'] = profit * reverse
    pos['p_rate'] = (1-price/eval_price)*100 *reverse
    pos['duration'] = pos['duration'] +1 if 'duration' in pos else 0
    if 'max_down_rate' in pos:
      if pos['max_down_rate'] > pos['p_rate']:
         pos['max_down_rate'] = pos['p_rate']
    else:
        pos['max_down_rate'] = pos['p_rate']
    pos['commit'] = False
    #print("再計算:",pos)

def calc_position_sum(idx):  
  #ポジションのサマリーを記録する。
  profit_sum = 0
  for pos in positions:
    profit_sum += pos['profit']
  ps ={
      'date': idx.strftime('%Y-%m-%d'),
      'profit_sum': profit_sum
  }
  global df_positions_sum_log 
  df_positions_sum_log = df_positions_sum_log.append(ps,ignore_index=True)

#-------- パラメータの変更(実験) -------
def change_param():
  buy_sum = 0
  sell_sum = 0
  for pos in positions:
    if(pos['type'] == "buy"):
      buy_sum += pos['profit']
    else: 
      sell_sum += pos['profit']
  print("date, buy_sum,sell_sum:",_ROW.name, buy_sum,sell_sum)
  global is_max_buy
  global is_min_buy
  if sell_sum < sell_danger_line:
    is_max_buy = False
    is_min_buy = False
    print("@@@@@@@buy parameter changed @@@@@@@@@@@@@@")
  else:
    is_max_buy = True
    is_min_buy = True
  if buy_sum < buy_danger_line:
    is_max_sell = False
    is_min_sell = False
    print("@@@@@@sell parameter changed@@@@@@@@@@@@@@@")
  else:
    is_max_sell = True
    is_min_sell = True

  #--- ma75 で判断----

  if _ROW['ma75_rate'] > 10:
      is_max_buy = False
      is_min_buy = False
  elif _ROW['ma75_rate'] <-10:
      is_max_sell = False
      is_min_sell = False
  else:
      is_max_buy = True
      is_min_buy = True
      is_max_sell = True
      is_min_sell = True



#-------- ポジションの決済　--------
def check_available_commit(deal_type):
  return True
  #--- なんかうまくいかない
  deal_count = 0
  for pos in positions:
    if pos['type'] == deal_type:
      deal_count += 1

  if len(positions)>0 and deal_count==1:
    return False
  return True

def commit(pos):
  positions.remove(pos)
  pos['date'] = _ROW.name  #緊急対応

  global df_commits 
  df_commits = df_commits.append(pos,ignore_index=True)

def commit_by_batch():

  pos_l = [pos for pos in positions if pos['commit']]

  for pos in pos_l:
    if pos['commit']:
      if pos['profit'] < 0:
        print("-相殺実行:",pos)
        #positions.remove(pos)
        commit(pos)
      else:
        print("+相殺実行",pos) 
        #positions.remove(pos)
        commit(pos)

def setoff_commit():
  plus_count = 0
  plus_price = 0
  is_setoff = False
  for pos in positions:
    if pos['profit'] > 0:
      plus_count += 1
      plus_price += pos['profit'] 
  if plus_count >= 2:
    print("相殺決済可能:",plus_price)
    #ここで相殺ポジションを一時的に昇順に並び替えたい
    #sorted_positions = sorted(positions, key=lambda x: x['profit'])
    #for idx,pos in enumerate(sorted_positions):
    for idx,pos in enumerate(positions):
      if pos['profit']<0 and abs(pos['profit']) < plus_price:
        #print("-相殺実行:")
        pos['commit'] = True
        #commit_by_index(idx)
        is_setoff = True
        break
    if is_setoff:
      for idx,pos in enumerate(positions):
        if pos['profit'] > 0:
          #print("＋相殺実行:")
          pos['commit'] = True
      [print(pos) for pos  in positions]
      commit_by_batch()
      return True

  return False

def commit_by_index(idx):
  pos = positions[idx]
  print("\ncommit:",pos)
  commit(pos)

def commit_by_index_list(idx_l):
  for idx in idx_l:
    pos = positions[idx]
    print("\ncommit:",pos)
    commit(pos)

def commit_position():
  print(len(positions),end=",")
  if setoff_commit():
    pass
    #return True
  #--- 通常決済 ----
  for i, pos in enumerate(positions):
    if pos['p_rate'] > commit_rate:
      if check_available_commit(pos['type']):
        commit_by_index(i)
        break

  #--- 強制ロスカット ----
  for i, pos in enumerate(positions):
    if pos['p_rate'] < loss_cut_rate:
      if check_available_commit(pos['type']):
        commit_by_index(i)
        break

  #--- 時限ロスカット ----
  for i, pos in enumerate(positions):
    if pos['duration'] > max_duration:
      if check_available_commit(pos['type']):
        commit_by_index(i)
        break

  return True

#-------- 取引シミュレーション -----
#ticker_info = yf.Ticker("5020.T")
ticker_info = yf.Ticker(stocker)
#hist = ticker_info.history(start=deal_start,end=deal_end)
hist = ticker_info.history(period="max")

if hist.isna()['Close'][-1:].bool():
  hist= hist[:-1] #最終行のみ削除




hist['ma5'] = hist['Close'].rolling(window=5).mean()
hist['ma25'] = hist['Close'].rolling(window=25).mean()
hist['ma25_rate'] = (hist['Close']-hist['Close'].rolling(window=25).mean())/hist['Close']*100
hist['ma75'] = hist['Close'].rolling(window=75).mean()
hist['ma75_rate'] = (hist['Close']-hist['Close'].rolling(window=75).mean())/hist['Close']*100

hist = hist.query(f"index >= '{deal_start}' and index <= '{deal_end}'")

for idx,row in hist.iterrows():
#for idx,row in hist.iterrows():
  _ROW = row
  calc_position(idx,row)
  change_param()
  deal_type = eval_position(row)
  if deal_type:
      entry_position(idx,row,deal_type)
  calc_position(idx,row)
  commit_position()
  calc_position_sum(idx)

print("\ncommits:")
#----- 最終計
commit_total = 0
df_commits['cumsum'] = df_commits['profit'].cumsum()
pd.set_option('display.expand_frame_repr', False)
print(df_commits)
commit_total = int(df_commits['profit'].sum())
print("commi total:",commit_total, df_commits['profit'].count())

print("\npositions:")
position_total = 0
for position in positions:
  print("psition:",position)
  position_total += position['profit']
print("psition total:",position_total, len(positions))
print("最大/最小:",df_positions_sum_log['profit_sum'].max(),df_positions_sum_log['profit_sum'].min())


print("\nfinal:", commit_total + position_total)

plt.plot(hist['Close'])

```
